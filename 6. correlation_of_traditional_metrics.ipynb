```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr 

LANG = 'FR'
model = 'PG'
#MAIN = f'resulting_data/{LANG}_latin/'
MAIN = f'resulting_data/{LANG}/'

path_trad = MAIN+f'{model}_traditional.csv'

# BERT dan PG 
path_human_focus = f'mturk/annotation_result/{LANG}/human_focus_final.csv'
path_human_coverage = f'mturk/annotation_result/{LANG}/human_coverage_final.csv'

def eval (h, c):
    for col in c.columns[1:]:
        score = pearsonr(h, c[col])[0]
        print(col+','+str(score))
    print()

trad = pd.read_csv(path_trad)
human_foc = pd.read_csv(path_human_focus)
human_cov = pd.read_csv(path_human_coverage)

human_foc=human_foc[human_foc['model']==model]
human_cov=human_cov[human_cov['model']==model]

data_trad = {}
for idx, row in trad.iterrows():
    data_trad[row['id']] = row
```

```python
#focus combined
filtered_human = []
filtered_trad = pd.DataFrame(columns=trad.columns)
for idx, row in human_foc.iterrows():
    filtered_human.append(row['score'])
    filtered_trad = filtered_trad.append(trad[trad['id']==row['id']])
print('Focus', model, MAIN, len(filtered_human))
eval(filtered_human, filtered_trad)

#coveraged combined
filtered_human = []
filtered_trad = pd.DataFrame(columns=trad.columns)
for idx, row in human_cov.iterrows():
    filtered_human.append(row['score'])
    filtered_trad = filtered_trad.append(trad[trad['id']==row['id']])
print('Coverage', model, MAIN, len(filtered_human))
eval(filtered_human, filtered_trad)
```
